```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Imports.

```python
url = "https://raw.githubusercontent.com/girishkuniyal/Predict-housing-prices-in-Portland/master/ex1data2.txt"
data= pd.read_csv(url, header=None, names=['size','beds','price'])
```

Same data used as lecture.

```python
df = data[['size','beds', 'price']]
df.head(47)
```

Here we have the data frame we will need for multiple variable regression.

```python
x = df[['size', 'beds']].values
x_graph = df['size'].values
y = df['price'].values
x = x.reshape(-1,2)
y = y.reshape(-1,1)
print(x)
```

Transfer the data into a matrix and vector respectively.

```python
reg_false = LinearRegression(fit_intercept=False).fit(x, y)
reg_true = LinearRegression(fit_intercept=True).fit(x, y)
```

Perform both regressions.

```python
reg_false.score(x, y)
```

```python
reg_true.score(x,y)
```

Find the scores.

```python
reg_false.coef_
```

```python
reg_true.coef_
```

```python
reg_true.intercept_
```

```python
reg_false.predict(np.array([[2980, 3]]))
```

```python
reg_true.predict(np.array([[2980, 3]]))
```

```python
plt.scatter(x_graph, y, color = "red")
plt.plot(x, reg_false.predict(x), color = "green")
plt.title("Int False regression")
plt.xlabel("Features")
plt.ylabel("Price")
plt.show()
```

```python
plt.scatter(x_graph, y, color = "blue")
plt.plot(x, reg_true.predict(x), color = "green")
plt.title("Int True regression")
plt.xlabel("Features")
plt.ylabel("Price")
plt.show()
```
